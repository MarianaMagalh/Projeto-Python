Analise de Dados do Programa de Itiner√°rios Formativos(PROIF)




```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

arquivo = pd.read_csv('PROIF_Relatorio_Status.csv', sep=',')
```

```python
uf = arquivo['UF']
escolas = arquivo['TOTAL_ESCOLA']

plt.figure(figsize=(10,6))
plt.plot(uf, escolas, marker='o', linestyle='-', color='b')
plt.title('Quantidade de escolas-PROIF por estados')
plt.xlabel('UF', )
plt.ylabel('Escolas')
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
uf = arquivo['UF']
matriculas = arquivo['MATRICULAS']

plt.figure(figsize=(10,6))
plt.bar(uf, matriculas)

plt.title('Quantidade de matriculas por estados')
plt.ylabel('Matriculas')
plt.xlabel('Estados')

plt.gca().yaxis.set_major_locator(ticker.MaxNLocator(integer=True))
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

```python
verba = arquivo['VERBA']
escolas = arquivo['TOTAL_ESCOLA']
uf = arquivo['UF']

plt.figure(figsize=(10,6))
plt.bar(uf, escolas, color='lightgreen')
plt.bar([x + 0.2 for x in range(len(uf))], escolas, color='salmon')

plt.title('Verba por escola')
plt.ylabel('Verba')
plt.xlabel('Escolas(estados)')
plt.legend(['estados', 'escolas'])
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```
