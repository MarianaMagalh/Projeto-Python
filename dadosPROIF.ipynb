Analise de Dados do Program de Itinerários Formativos(PROIF)

O arquivo "PROIF_Relatorio_Status.csv", mostra os dados sobre as adesões das Unidades Escolares via Programa Dinheiro Direto na Escola(PDDE) no ano de 2021, comode repasses finaceiros e matrículas de estudantes em escolas participantes do programa.


Na análise realizada, foi visto o número de escolas participantes, por UF, á quantidade de matrículas(estudantes) e recursos financeiros repassados por escolas participante, sendo totalizando por UF.


```python
# Aqui estou apenas mostrando as informação da tabela

import pandas as pd
import matplotlib.pyplot as plt

arquivo = pd.read_csv('PROIF_Relatorio_Status.csv', sep=',', encoding='latin-1')

print(arquivo.head())
print(arquivo.columns)
```

```python
# 

uf = arquivo['UF']
escolas = arquivo['TOTAL_ESCOLA']

plt.plot(uf, escolas, marker='o', linestyle='-', color='b')
plt.xlabel('UF')
plt.ylabel('Escolas')
plt.title('Gráfico de Linha')
plt.grid(True)
plt.show()
```
